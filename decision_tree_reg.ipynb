```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

dataset =pd.read_csv('train-data-1-_1_.csv')

dataset =dataset.fillna(dataset.mean())

dataset.drop(["Name", "Year"], axis = 1, inplace = True) 

x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values


#from sklearn.impute import SimpleImputer
#imputer =SimpleImputer(missing_values =np.nan,strategy='mean')


from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct =ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
x=np.array(ct.fit_transform(x))


from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()

x[:,14] =le.fit_transform(x[:,14])
x[:,12] =le.fit_transform(x[:,12])
x[:,13] =le.fit_transform(x[:,13])



from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=0)


from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train[:,1] =sc.fit_transform(x_train[:,1].reshape(1,-1))
x_test[:,1] =sc.fit_transform(x_test[:,1].reshape(1,-1))
x_train[:,11] =sc.fit_transform(x_train[:,11].reshape(1,-1))
x_test[:,11] =sc.fit_transform(x_test[:,11].reshape(1,-1))

print(x_train[:,11])


from sklearn.tree import DecisionTreeRegressor
regressor =DecisionTreeRegressor(random_state=0)
regressor.fit(x_train,y_train) 

y_pred=regressor.predict(x_test)


from sklearn.metrics import r2_score
print(r2_score(y_pred,y_test))


```
